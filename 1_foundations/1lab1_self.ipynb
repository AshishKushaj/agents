```python
from openai import OpenAI


openai = OpenAI()
```

```python
message1=[{"role":"user","content":"What is 8*9?"}]

response1=openai.chat.completions.create(
    model="gpt-5-nano",
    messages=message1
)
```

```python
print(response1.choices[0].message.content)
```

```python
question2 = "Please propose a hard, challenging question to assess someone's IQ. Respond only with the question."
message2=[{"role":"user","content":question2}]

response2=openai.chat.completions.create(
    model="gpt-5-mini",
    messages=message2
)
```

```python
question = response2.choices[0].message.content
print(question)
```

```python
message3=[{"role":"user","content":question}]

response3=openai.chat.completions.create(
    model="gpt-5-mini",
    messages=message3
)
```

```python
answer = response3.choices[0].message.content

print(answer)
```

```python
q1 = "pick a business area that might be worth exploring for an agentic ai opportunity"

mess1 = [{"role":"user","content":q1}]

res1 = openai.chat.completions.create(
    model="gpt-5-mini",
    messages=mess1
)
```

```python
ans1 = res1.choices[0].message.content

print(ans1)
```

```python
ans1_summary = "give brief summary of the following: " + ans1
mess_summary = [{"role":"user","content":ans1_summary}]


response_summary = openai.chat.completions.create(
    model="gpt-5-nano",
    messages=mess_summary
)
```

```python

print(response_summary.choices[0].message.content)
```

```python
q2 = "give a pain pointt in the following industry that might be worth exploring for an agentic ai opportunity : " + response_summary.choices[0].message.content

mess2 = [{"role":"user", "content":q2}]

res2 = openai.chat.completions.create(
    model="gpt-5-mini",
    messages=mess2
)
```

```python
ans2 = res2.choices[0].message.content
print(res2.choices[0].message.content)
```

```python


q2_summary = "summerize the following: " + ans2

mess_summary2 = [{"role":"user","content":q2_summary}]

response_summary2 = openai.chat.completions.create(
    model="gpt-5-nano", 
    messages=mess_summary2
)
```

```python
q2_summarized = response_summary2.choices[0].message.content
print(q2_summarized)
```

```python

problem = [{"role":"user" , "content":q2_summarized}]

res_sol =  openai.chat.completions.create(
    model="gpt-5-mini", 
    messages=problem
)
```

```python

solution = res_sol.choices[0].message.content

print(solution)
```
